```python
from FD import FD
from tcand import minimal_core_ip_exact, fd_closure

import random
import string
from itertools import combinations
from typing import List, Set
```

```python
def make_vars(n):
    return [f"X{i}" for i in range(n)]

def check_solution(fds, sol, target):
    clo = fd_closure(fds, sol)
    return target <= clo

def print_solution(label, fds, sol, target):
    print(f"{label}: {sol} | size={len(sol)} | valid={check_solution(fds, sol, target)}")

def demo_basic():
    print("=== Demo 1: Canonical example ===")
    fds = [
        FD({"A"}, {"B"}),
        FD({"B"}, {"C"}),
        FD({"A", "D"}, {"E"}),
    ]
    print(fds)
    target = {"A", "B", "C", "D", "E"}

    print("Target:", target)
    print("Closure({A,D}):", fd_closure(fds, {"A", "D"}))

    sol = minimal_core_ip_exact(fds, target)
    print("IP solution:", sol)
    print()
```

```python
demo_basic()
```
